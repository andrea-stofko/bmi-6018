**The goal of the following is to prepare a FastQC Report using the SeqIO module:**

```python
#import BioSeq
from Bio.Seq import Seq
```

```python
#import SeqUtils
from Bio import SeqUtils
```

```python
#import SeqIO (this will be used to analyze a sample fastq file)
from Bio import SeqIO
```

```python
#check the number of reads in mutant1 fastq file:
count = 0
for rec in SeqIO.parse("mutant_R1.fastq","fastq"): 
    count +=1
print(f"There are {count} reads in our sample")
```

```python
#view sequence data:
seq = []
for i in SeqIO.parse("mutant_R1.fastq","fastq"):
    seq.append(i.seq)
```

```python
seq
```

```python
#determine length of sequences:
seqLength = set([len(x) for x in seq]) #using set to remove duplicate values
print(seqLength)
```

```python
#import GC_fraction
from Bio.SeqUtils import gc_fraction
```

```python
#print GC content
seq_str = str(seq)
gc_content = gc_fraction(seq_str)*100
print(gc_content)
```

```python
#inspect the PHRED quality score of the reads:
mut1 = SeqIO.parse("mutant_R1.fastq","fastq")
mut1 = next(mut1)
mut1.letter_annotations['phred_quality']
```

```python
#removing reads that fall below a quality score of 20 (99% accurate):
cleaned_reads = (
    mut1
    for mut1 in SeqIO.parse("mutant_R1.fastq","fastq")
    if min(mut1.letter_annotations["phred_quality"]) >=20
)
count = SeqIO.write(cleaned_reads, "mutantR1cleaned.fastq","fastq")
print(f"New number of reads is {count}")
```
