```python
import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
```

```python
#load and inspect arrythmia data
a_data = pd.read_csv('arrhythmia.data', header = None, usecols = [0,3,168]) #using age, weight, and qrsta measurement
a_data.head()
```

```python
#drop any ?'s and NAs
a_data.replace('?', np.nan, inplace=True)
a_data = a_data.dropna()
print(len(a_data)) #length of dataset did not change 
```

```python
#rename columns
a_data = a_data.rename(columns = {0:'Age', 3:'Weight', 168:'QRSTA'}) 
a_data.head()
```

```python
#view mean, std, etc prior to scaling
a_data.describe()
```

```python
#visualizing the data
plt.figure(figsize=(10, 6))
sns.scatterplot(data = a_data, x = 'Age', y = 'Weight', hue = 'QRSTA')
plt.show()

''' Not sure that I picked the best columns to analyze based on the way data clusters below. 
    I was expecting clearer separation between groups. However, I wasn't completely sure what 
    the best method was to determine the columns that should be used for this'''
```

```python
print(a_data[['Age', 'Weight', 'QRSTA']].dtypes) #double-checking data types before scaling
```

```python
scaler = StandardScaler() 
```

```python
#use scaler to normalize data so that mean is 0 and std is 1
a_data[['Age_T', 'Weight_T', 'QRSTA_T']] = scaler.fit_transform(a_data[['Age', 'Weight', 'QRSTA']])
```

```python
#view scaled data
a_data
```

```python
#optimizie number of clusters using function:

def optimize_k_clusters(data, max_k):
    means = []
    inertias = []

    for k in range (1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        means.append(k)
        inertias.append(kmeans.inertia_)

    #create elbow plot
    fig = plt.subplots(figsize=(10,6))
    plt.plot(means, inertias)
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show
```

```python
#optimal k means clusters using just two of the variables (it seemed that most example models recommended only 2)
optimize_k_clusters(a_data[['Age_T','Weight_T']],10)
```

```python
'''Based on this graph and where the 'elbow' occurs, I would determine that 4 is the optimal number
   of clusters to use for this subset of the data'''
```

```python
#assign k means clusters value
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(a_data[['Age_T', 'Weight_T']])
```

```python
#create new column for each assigned cluster
a_data['Cluster']=kmeans.labels_
a_data
```

```python
#plot new cluster results
sns.scatterplot(data = a_data, x = 'Age_T', y = 'Weight_T', hue = 'Cluster')
plt.show()
```
