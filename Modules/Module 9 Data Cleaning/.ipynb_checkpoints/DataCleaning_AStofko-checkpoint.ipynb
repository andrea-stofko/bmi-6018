```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#import and view the flights data
flights_data = pd.read_csv('flights.csv')
flights_data.head(2)
```

```python
#inspecting NA values for each column:
print(flights_data.isna().sum())
```

```python
# Percentage of missing data for each column 
percent_of_missing_data = flights_data.isnull().sum() / len(flights_data) * 100
percent_of_missing_data
#this does not return a high percentage, leave columns
```

```python
#removing any duplicates:
flights_data.drop_duplicates(inplace=True)
flights_data.count() #checking for total number of flights = 336,776 (this did not change)
```

```python
#Question 1 How many flights were there from JFK to SLC? Int
jfk = flights_data.loc[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC')]
jfk.head()
```

```python
q1=len(jfk)
print('Q1: Number of flights from JFK to SLC:',q1)
print(type(q1))
```

```python
#Question 2 How many airlines fly to SLC? Should be int

#filter original flights data so that destination = SLC
slc = flights_data.loc[flights_data['dest']=='SLC']
#slc.head()

#create subset that is only the carriers that go to SLC
slc_carriers = slc['carrier']
slc_carriers.head(1)
      
q2 = slc.carrier.nunique()
print("Q2: The number of airlines that fly into SLC is",q2,"and it is",type(q2))
```

```python
#Question 3 What is the average arrival delay for flights to RDU? float

#filter only flights that go to RDU
rdu = flights_data[flights_data['dest']=='RDU']
#rdu.head(1)

#calculate and print mean
q3 = rdu['arr_delay'].mean()
print("Q3: The average arrival delay for RDU is:",q3,"and the class type is:",type(q3))
```

```python
#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float

#ID total number of Seattle flights:
sea = flights_data[flights_data['dest']=='SEA']
sea.head(1)
```

```python
#find total number of flights that go to Seattle
sea_counts = sea['flight'].count()
print(sea_counts)
#total number of seattle flights = 3923
```

```python
#find total number of flights from NY airports: LGA or JFK
#ny_sea = sea[(sea['origin'] =='LGA') | (flights_data['origin'] =='JFK')]
#ny_sea.head(1)
ny_sea_counts = ny_sea['flight'].count()
print(ny_sea_counts) #there are 2092 flights from NY to SEA
```

```python
#what is the proportion?
q4 = ny_sea_counts/sea_counts
print('Q4: The percentage of flights to Seattle that are from NY airports is',q4*100,"%")
```

```python
#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)

#add in date column
flights_data['date']=pd.to_datetime(flights_data[['year','month','day']]) 
#flights_data.head()

flights_data['date'] = flights_data['date'].dt.strftime('%Y/%m/%d')
flights_data.head()
```

```python
#find mean for departure delays
dep_delays = flights_data.groupby('date')['dep_delay'].mean()
```

```python
#Question 5 Answer
q5 = dep_delays.sort_values(ascending = False)
print('Q5: The date with the longest departure delays is', q5.head(1))
```

```python
#Question 6 Which date has the largest average arrival delay? pd slice with date and float
arr_delays = flights_data.groupby('date')['arr_delay'].mean()
```

```python
q6 = arr_delays.sort_values(ascending = False)
print('Q6: The date with the longest arrival delays is', q6.head(1))
```

```python
#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime

#create subset for departures from LGA or JFK
ny_flights = flights_data.loc[((flights_data['origin']=='LGA') | (['origin']=='JFK')) & (flights_data['year']==2013)]
ny_flights.head(2)
```

```python
#add new column for speed
ny_flights.loc[:,'speed']=ny_flights['distance']/ny_flights['air_time']
ny_flights.head(2)
```

```python
#Question 7 answer: Which flight departing NY flew the fastest?
q7 = ny_flights[['tailnum','speed']].sort_values(by='speed', ascending=False)
print('Q7:The flight that flew the fastest from LGA or JFK in 2013 is:\n',q7.head(1))
```

```python
#Question 9 How many observations were made in Feburary? Int

#convert weather_data with no NAs to numpy array
weather_data_np = q8.to_numpy()
#print(weather_data_np)

#select for all rows where where month = 2 (fourth column)
feb_weather = weather_data_np[weather_data_np[:,3] == 2.0] 
#print(feb_weather)

print("Q9: The number of observations in February is",len(feb_weather))
print(type(len(feb_weather)))
```

```python
#Question 10 What was the mean for humidity in February? Float
q10 = np.mean(feb_weather[:,8])
print("Q10: The average humidity in February is:",q10)
print(type(q10))
```

```python
#Question 11 What was the std for humidity in February? Float
q11 = np.std(feb_weather[:,8])
print("Q11: The standard deviation in humidity in February is:",q11)
print(type(q11))
```
