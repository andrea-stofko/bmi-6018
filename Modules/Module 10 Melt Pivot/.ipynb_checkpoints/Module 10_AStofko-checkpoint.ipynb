```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#import data
bc = pd.read_csv("breast-cancer.data", sep=',', header = None)
bc.head()
```

```python
#assign column names:
bc.rename(columns = {0:'class',
                     1:'age',
                     2:'menopause',
                     3:'tumorSize',
                     4:'invNodes',
                     5: 'nodeCaps',
                     6:'degMalig',
                     7:'breastLR',
                     8:'breastQuad',
                     9:'irradiat'},
          inplace = True)
bc.head()
```

```python
#check for missing data (it is noted in the .names file that Nulls are replaced with '?'
print(bc.isna().sum())
```

```python
#barplot with original data
sns.barplot(data = bc, x='menopause', y='degMalig', hue = 'age') #looking at degree of malignancy based on age and menopause status
```

```python
##MELT##
```

```python
#using melt to reshape df
bc_melted = bc.melt(id_vars='class')
bc_melted.head()
```

```python
##PIVOT##
```

```python
#using pivot to make the dataframe the same as the original
bc_pivoted = bc_melted.pivot_table(index='class',columns ='variable', values='value', aggfunc='first') 
bc_pivoted.reset_index(inplace=True)

bc_pivoted.head()
```

```python
#renaming columns
bc_pivoted.columns = ['class', 'age', 'breastLR', 'breastQuad', 'degMalig', 'invNodes', 'irradiat', 'menopause', 'nodeCaps','tumorSize']
bc_pivoted.head()
```

```python
##AGGREGATION##
```

```python
#what is the the maximum degreee of malignancy? (using aggregation)
maxMalig= bc['degMalig'].max()
print(maxMalig)
```

```python
#what is the average degree of malignancy? (using aggregation)
avgMalig = round(bc['degMalig'].mean(),2)
print(avgMalig)
```

```python
##GROUPBY##
```

```python
#What is the average degree of malignancy when grouping by age?
MaligByAge = round(bc.groupby('age')['degMalig'].mean(),2)
print(MaligByAge) 
```

```python
#applying custom aggregation and squaring degMalig 
def malignancy(x):
    return x**2
bc.groupby(['menopause'], as_index = False)['degMalig'].apply(malignancy)
```

```python
#importing long beach VA heart disease data (dataset with more integers)
hd = pd.read_csv("processed.va.data", sep=',', header = None)
hd.head() #importing long beach VA heart disease data (dataset with more integers)
```

```python
#assign column names:
hd.rename(columns = {0:'age',
                     1:'sex',
                     2:'cp',
                     3:'trestbps',
                     4:'chol',
                     5: 'fbs',
                     6:'restecg',
                     7:'thalach',
                     8:'exang',
                     9:'oldpeak',
                    10:'slope',
                    11:'ca',
                    12:'tha',
                    13:'dx'},
          inplace = True)
hd.head()
```

```python
#replace ?'s with NaN in thalach column
hd['thalach'] = hd['thalach'].replace('?', np.nan)
hd.head()
```

```python
#iterate over and create target max column
targetMax = []
for index, row in hd.iterrows():
    targetMax.append(220-row['age'])

hd['targetMax'] = targetMax
hd.head()
```

```python
#print(hd['thalach'].dtype) #object

#change to int
hd['thalach'] = hd['thalach'].astype('Int64')
```

```python
#calculate the difference between target and max HR
differenceinHR = []
for index, row in hd.iterrows():
    differenceinHR.append(row['targetMax']-row['thalach'])

hd['differenceinHR'] = differenceinHR
hd.head()
```
